```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_') 
df = df.rename(columns = {'purchase_amount_(usd)': 'purchase_amount'})
# we use this two function to change the column name into lower characters because when we apply some python and sql operation onto the data
# we have a mistake like column name type we forgot some space b\w the columns name so these function are use to read the column name easy 
# and perform some opration without any querry or problem ..
```

```python
# Create new column age_group
labels = ['Young Adult','Adult','Middle-Age','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels= labels)
# the "qcut " function is used for gave the spacific age to a category like above 60 gave the senior etc which  is used to easy to understanding
# in the marketinng puspose and for the people.
```

```python
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90

}

df['purchase_frequency_date']  = df['frequency_of_purchases'].map(frequency_mapping)

# first we created a distinory called frequency mapping that maps each text frequency to the equalent number of days.
# Then we use the map function to replace the text column with the crossponding numbers of days.
```
